# Summarizing Distributions

## Probability VS. Statistics
**Probability** says given how data are generated, what is the probability of a particular event? It is forward direction, from model to data. In other words we know the process and we ask what happens? Therefore, going from cause to effect. On the other hand, **statistics** asks what probability distribution could have generated this data? It is backward direction, from data or observations to model. We see data and we ask what caused it? Hence, going from effect to cause.

## Summary Features of Random Variables
We begin by defining measures of the *"center"* and *"spread"* of the distribution of a single random variable.

In the real world, the full probability distribution of a random variable maybe highly complex and therefore may not have a simple mathematical representation. Hence, we will want to be able to describe at least certain key features of such distributions nonparametrically, that is, without assuming that these distributions can be fully characterized by a distribution function with a finite number of parameters.

### Expected Values
The expected value also known as expectation or mean of a random variable can be thought of as the value we would obtain if we took the average over many, many realizations of a random variable. It is the most commonly used measure of the "center" of a probability distribution.

Expected value of a discrete random variable:

$$E[X] = \sum x f(x) $$

Expected value of a continuous random variable:

$$E[x] = \int_{-\infty}^{\infty} x f(x)$$

**Note:** The expected value is an operator which takes on a random variable as an input and returns a number. So, when we compute the expected value of a random variable, we are applying the expectation operator.

### Law of the Unthinking Statistician (LOTUS)
To find the expected value of a function of a random variable, you don't need the distribution of the function itself. Just use the original variable. In other words, instead of first transforming your data, then figuring out the new distribution, you can skip all that and just apply the function directly inside the expectation. The reason it is called **"unconscious"** is because we don't think about the new distribution at all.

*Want the average of a transformed variable? Just apply the transformation and average using the original probabilities*

### Expectation of a Function of Two Random Variables

#### Discrete Case
If \(X\) and \(Y\) are discrete random variables:

$$
\mathbb{E}[h(X, Y)] = \sum_{x} \sum_{y} h(x, y)\, f_{X,Y}(x, y)
$$

where $f_{X,Y}(x, y)$ is the joint probability mass function.

---

#### Continuous Case
If \(X\) and \(Y\) are continuous random variables:

$$
\mathbb{E}[h(X, Y)] = \int \int h(x, y)\, f_{X,Y}(x, y)\, dx\, dy
$$

where $f_{X,Y}(x, y)$ is the joint probability density function.

---

### Linearity of Expectation

For any constants $a, b$ and random variables $X, Y$:

$$
\mathbb{E}[aX + bY + c] = a\,\mathbb{E}[X] + b\,\mathbb{E}[Y] + c
$$

**Important:** Linearity holds **even if the variables are dependent**.
